```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
h = 0.05
k = 195
k_list = [i for i in range(k)]

x = np.random.standard_normal(k)
x = np.append(x, [5, -4, 3.3, 2.99, -3])
x.sort()
```

```python
#task 1
mean = np.mean(x)
var = np.var(x)
isOutlying = lambda x: np.abs(x - mean) > 3*var
outlyings = [val for val in x if isOutlying(val)]
outlyingsIndexes = [i for i, val in enumerate(x) if isOutlying(val)]
normIndexes = [i for i, val in enumerate(x) if not isOutlying(val)]
x_d = np.delete(x, outlyingsIndexes)
print("Выбросы: ", outlyings)
plt.plot(normIndexes, x_d, '*', label = 'x')
plt.plot(outlyingsIndexes, outlyings, 'o', label = 'Выбросы')
plt.grid()
plt.legend()
plt.title("Выбросы по 3$\sigma$")
plt.show()
```

```python
outlyings = plt.boxplot(x, vert=False)
print("Выбросы по Тьюки: ", [x.get_xdata() for x in outlyings["fliers"]])
plt.title("Боксплот Тьюки")
plt.show()


```
